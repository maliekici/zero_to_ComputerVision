#**Color Spaces**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def image_show1(title = "Title", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('img_folder/1.jpg')
#split
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
image_show1("Blue Channle", B, 4)
image_show1("Blue Channle", G, 4)
image_show1("Blue Channle", R, 4)
```

```python
#image shape with, hight,(500x500) dimension we will take hight and with
zeros = np.zeros((image.shape[:2]), dtype="uint8") #image.shape[:2] = 500x500
image_show1("Red", cv2.merge([zeros, zeros, R])) 
image_show1("Green", cv2.merge([zeros, G, zeros]))
image_show1("Blue", cv2.merge([B, zeros, zeros]))
```

```python
image_show1("Merged", zeros, 3)
image_show1("Merged", zeros+100, 3)
```

```python
merged = cv2.merge([B, G, R])
image_show1("Merged", merged, 4)
```

```python
#boost blue color
merged2 = cv2.merge([B+100,G, R])
image_show1('boost blue', merged2, 5)
```

#**Drawing Images**

```python
def image_show2(title = "Title", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image_rgb = np.zeros((312, 312, 3), np.uint8)
image_gray = np.zeros((312, 312), np.uint8)
image_show2("Black Canvas -- RGB Color", image_rgb, 5)
image_show2("Black Canvas -- Gray Scale Color", image_gray, 5)
```

```python
#lets draw a line 
#note this is an inplace operation, meaning it chances the input image
#unlike many other OpenCV functions that returns a new image leaving the input unaffected
#remember our image was the black canvar
#cv2.line(image, start cordinate, stop cordinate, color, thickness)
cv2.line(image_rgb, (10,10), (311,311), (255, 127, 0), 5)
image_show2("line of a our black canvas", image_rgb, 5)
```

```python
#drawing rectangles
#so is this an inplace operation, meaning it chances the input image
#cv2.line(image, start cordinate, stop cordinate, color, thickness)
image_rgb2 = np.zeros((312, 312, 3), np.uint8)
cv2.rectangle(image_rgb2, (100,100), (300, 250), (127, 50, 127), 10)
image_show2("black canvar with pink rectangle",image_rgb2, 5)
```

```python
#drawin circle
#cv2.circle(image, center, radius, color, fill)
image_rgb3 = np.zeros((312,312,3), np.uint8)
cv2.circle(image_rgb3, (100,100), 50, (15,150,50), -1)
image_show2("black canvar with circle", image_rgb3, 5)
```

```python
#drawing polygons
#polylines(img, pts, isClosed, color, thickness) pts = points this is our point array
image_rgb4 = np.zeros((312,312,3), np.uint8)
array_list = np.array(  [[10,50], [200,50], [90,200], [50,200]], np.int32  )
array_list = array_list.reshape((-1,1,2))
cv2.polylines(image_rgb4, [array_list], True, (0,0,255), 3)
image_show2("Black canvas with polygons", image_rgb4, 5)
```

#**Adding Text with cv2.putText**

* cv2.putText(image, 'Text to Display', bottom left ST, font, font size, Color, Thickness
* you can access available font list in google 

```python
def image_show3(title = "Title", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image_rgb5 = np.zeros((512, 512, 3), np.uint8)
ourString = "hello world"
cv2.putText(image_rgb5, ourString, (155,290), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 3, (40,200,0), 4)
image_show3("text",image_rgb5, 5)
```
