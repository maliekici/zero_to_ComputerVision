```python
image = cv2.imread('./img_data/1.jpg')
# cv2.imshow("image", image) #if you are working with the IDE you downloaded to your computer, this command will probably work for you.
```

```python
#displaying images
from matplotlib import pyplot as plt
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
def image_show(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
#lets call our function
image3 = cv2.imread('img_data/3.jpg')
image_show("display", image3)
```

```python
#saving image
cv2.imwrite("outputimg3.jpg", image3) #or .png
```

```python
#displaying img dimension remember images are arrays.
import numpy as np
print(image3.shape) #open cv stores images as np obejcts
```

```python
img_src = cv2.imread('img_data/3.jpg',0)
print(img_src)
```

```python
#grayscaling images
def image_show2(title = "Image", image = None, size = 10):
    width, high = image.shape[0], image.shape[1]
    aspect_ratio = width / high
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image_show2("Title", image3, size=5) #if you reduce the size value foto will smaller
```

```python
grey_image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
image_show2("Gray Image", grey_image3, size = 4)
```
